```python

from nltk.tokenize import word_tokenize
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt
import PyPDF2
import numpy as np
import os
from PIL import Image
```

```python
filepath = '/git/emileDesmaili.github.io/files/CV.pdf'
# creating a pdf file object
pdfFileObj = open(filepath, 'rb')
  
# creating a pdf reader object
reader = PyPDF2.PdfReader(pdfFileObj)
  
# printing number of pages in pdf file
print(len(reader.pages))
  
# creating a page object
pageObj = reader.pages[0]
  
# extracting text from page
text = pageObj.extract_text()
  
# closing the pdf file object
pdfFileObj.close()

text
```

```python
mask = np.array(Image.open("images/mask.png"))
font_path = 'assets/fonts/windfonts/impact.ttf'
stopwords = STOPWORDS 
#add custom list of stopwords
stopwords.update(['Puma','Michael','New','York','City','NYC','NY','USA','United','States','America','US',
                  'American','Upmanu','Lall','Francis','M','Eva','email',
                  'Jan','Emile','Esmaili', 'S','J','Ludlow','Jobbova','Ehsan','Aric', 'Cutuli','Azhar','Muhammad',
                  'Rachata','Muneepeerakul', 'Early','Andrew','B','Singh','Universit'])

# Generate the wordcloud
wordcloud = WordCloud(background_color="white", stopwords=STOPWORDS, font_path=font_path,
                       max_words=500, mask=mask,
                         contour_width=0, contour_color='steelblue', colormap='Reds').generate(text)

# Display the wordcloud image
plt.figure(figsize=(6,8), facecolor=None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad=0)
plt.show()
```

```python
wordcloud.to_file('images/wordcloud.png')
```
